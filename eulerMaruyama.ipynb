```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
```

This is a generalisation of the regular Euler method (for ODEs) to stochastic differential equations

Consider the Ornstein-Uhlenbeck process:

$dY_t=\theta (\mu - Y_t)dt+\sigma dW_t$

where

$Y_t$ is the stochastic process<br>
$dt$ is the time increment<br>
$dW_t$ is a Weiner process<br>

The stencil is
$Y_{n+1}=Y_n+a(Y_n)\Delta t + b(Y_n)\Delta W_n$

The initial condition is $Y_0 = 0$

```python
t0 = 3  # Starting time
T  = 7  # Final time
N = 1000  # Grid points
dt = float(T - t0) / N
y0 = 0

c_theta = 0.7
c_mu = 1.5
c_sigma = 0.06

def mu(y, t):
    """Implement the Ornstein–Uhlenbeck mu."""  # = \theta (\mu-Y_t)
    return c_theta * (c_mu - y)

def sigma(y, t):
    """Implement the Ornstein–Uhlenbeck sigma."""  # = \sigma
    return c_sigma

def dW(delta_t):
    """Sample a random number at each call."""
    return np.random.normal(loc=0.0, scale=np.sqrt(delta_t))

ts    = np.arange(t0, T, dt)
ys    = np.zeros(N)

ys[0] = y0

for _ in range(5):
    for i in range(1, ts.size):
        t = (i-1) * dt
        y = ys[i-1]
        ys[i] = y + mu(y, t) * dt + sigma(y, t) * dW(dt)
    plt.plot(ts, ys)

plt.show()
```
