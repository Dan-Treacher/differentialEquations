```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 15
```

Define the ODE as:

$u'=\mathrm{sin}(40\,t\,u)$

with initial value of 

$u(0)=1$

```python
# Define constants

K = 0.01  # Time step
xmin, xmax = 0, 1  # Domain
N = int(xmax / K)  # Number of points
t = np.linspace(xmin, xmax, N+1)  # Time axis
v = np.zeros_like(t)  # Preallocate the vector of solutions at the t steps

# Initial value
v[0] = 1
```

```python
# Runge-Kutta stages (One of the fourth order schemes)

for i in range(N):
    a = K*np.sin(40 * t[i] * v[i])  # RK stage 1
    b = K*np.sin(40 * (t[i]+K/2) * (v[i]+a/2))  # RK stage 2
    c = K*np.sin(40 * (t[i]+K/2) * (v[i]+b/2))  # RK stage 3
    d = K*np.sin(40 * (t[i]+K) * (v[i]+c))  # RK stage 4

    v[i+1] = v[i] + (1/6)*(a + 2*b + 2*c + d)  # Update next value based on the stages just calculated
```

```python
# Plot the solution

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(t, v)
ax.set_xlabel('t')
ax.set_ylabel('u(t)')
plt.show()
```
