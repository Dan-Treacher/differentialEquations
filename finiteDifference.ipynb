```python
# Import relevant libraries

import numpy as np
import matplotlib.pyplot as plt
#plt.rcParams.keys()
plt.rcParams['font.size'] = 15
```

Try first order ODE:

$\dfrac{dy}{dx}+A\left(1-x^3\right)y-x=0$

with the initial value of 

$y(0)=1$

```python
# Define equation and parameters

N = 50  # Number of points in the discretisation
delta = 1/N  # Spacing
A = 0.2  # Tuning constant
exponent = 3
IV = 1  # Initial value

x = np.linspace(0, 1, N)  # x axis
y_finiteDifference = np.zeros_like(x)  # y vector
y_finiteDifference[0] = IV  # Initial condition
```

```python
# Do the finite difference stepping

for i in range(1, len(x)):
    y_finiteDifference[i] = y_finiteDifference[i-1] - y_finiteDifference[i-1]*(A*(1-x[i]**3))*delta

y_analytic = np.exp(0.05*x**4 - 0.2*x)
# Note that this analytical result will be wrong if you've messed up the equation
```

```python
# Plot the finite difference result against the analytic one

plt.figure(1, figsize=(10,5))
plt.plot(y_finiteDifference, linewidth=3, label='Finite difference')
plt.plot(y_analytic, linewidth=3, label='Analytic')
plt.xlabel('x'), plt.ylabel('y'), plt.legend()
plt.show()
```
