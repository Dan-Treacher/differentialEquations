```python
# import required libaries

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 15
```

Define ODE as:

$\dfrac{du}{dt}=\mathrm{sin}(40\,t\,u)$

with initial condition of:

$u(0)=1$

```python
# Define constants

K = 0.01  # Time step
xmin, xmax = 0, 1  # Domain
N = int(xmax/K)  # Number of points
t = np.linspace(xmin, xmax, N+1)  # Time axis

# Preallocate the response variable(s) axes as rows of a matrix
v = np.zeros((3, N+1))
v[:,0] = 1  # Initial condition(s)
```

Note that because the higher order Adams-Bashforth schemes require more past functions values, we use the one step scheme to calculate the first few function values to act as initial conditions for the higher order ones.

```python
# Function to compute the Adams-Bashforth first order (Euler) solutions
def AB1(t, K):
    v = np.zeros_like(t)
    v[0] = 1  # Initial condition
    for i in range(len(t)-1):
        v[i+1] = v[i] + K*np.sin(40*t[i]*v[i])
    return v

initialValues = AB1(t[:4], K)  # These are the first few values according to the first order scheme
v[:,:len(initialValues)] = initialValues
```

```python
# Multistep methods

for i in range(N-3):
    v[0,i+1] = v[0,i] + K*np.sin(40*t[i]*v[0,i])  # Adams-Bashforth first order (Euler's scheme)
    v[1,i+2] = v[1,i+1] + (K/2)*(3*np.sin(40*t[i+1]*v[1,i+1]) - np.sin(40*t[i]*v[1,i]))  # Adams-Bashforth second order
    v[2,i+3] = v[2,i+2] + (K/12)*(23*np.sin(40*t[i+2]*v[2,i+2]) - 16*np.sin(40*t[i+1]*v[2,i+1]) + 5*np.sin(40*t[i]*v[2,i]))  # Adams-Bashforth second orde
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(t[:-3], v[:,:-3].T)
ax.set_xlabel('t'), ax.set_ylabel('u(t)')
plt.legend(('First order (Euler)', 'Second order', 'Third order'))
plt.show()
```
